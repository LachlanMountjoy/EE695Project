```python
import numpy as np
import pandas as pd

data = pd.read_csv('diabetes.csv')  
data.head()
```

```python
data.describe().transpose()
```

```python
data.info()
```

```python
data.hist(figsize = (15, 15), bins = 20)
```

```python
col_names=[]
for col in data.columns:
    col_names.append(col)
```

```python
import matplotlib.pyplot as plt
df=data.corr(method ='pearson')
df_array=df['Outcome']
df_array=df_array.to_numpy()
plt.figure(figsize=(15,10))
plt.bar(col_names[:8], df_array[:8], color ='#AA3589',
        width = 0.4)
for index, value in enumerate(df_array[:8]):
    plt.text(index, value+0.003,
             str("%.3f" %value),size=15)
plt.xlabel("Variables",size=20)
plt.ylabel("Correlation Values",size=20)
plt.title("Correlation Bar Plot",size=30);
```
